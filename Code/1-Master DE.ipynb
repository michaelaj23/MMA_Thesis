```python
# Function to convert time format to float representing minutes
def convert_to_minutes(time_str):
    if isinstance(time_str, float):
        return time_str  # Return the float value directly
    else:
        minutes, seconds = map(int, time_str.split(':'))
        return minutes + seconds / 60.0
```

```python
def find_duplicates_in_column(df, column_name):
    """
    Find and display a list of duplicates in a specified column of a DataFrame.

    Parameters:
    - df: DataFrame
        The DataFrame containing the column with potential duplicates.
    - column_name: str
        The name of the column to search for duplicates.

    Returns:
    - list
        A list containing unique duplicate values found in the specified column.
    """
    # Find duplicates in the specified column
    duplicate_values = df[df.duplicated(subset=[column_name], keep=False)]

    # Extract the list of duplicate values
    duplicate_list = duplicate_values[column_name].unique().tolist()

    return duplicate_list
```

```python
def convert_inches(height):
    """
    Convert height from feet'inches" format to inches.

    Parameters:
    - height: str
        Height value in the format feet'inches" (e.g., "5'10").

    Returns:
    - int
        Height in inches.
    """
    if pd.isna(height):  # Handling NaN values
        return height
    
    feet, inches = height.split("'")
    total_inches = int(feet) * 12 + int(inches.strip('"'))
    return total_inches
```

```python
def unique_column_values(dataframe, column_name):
    """
    Returns a list of unique values in the specified column of a DataFrame.
    
    Args:
    - dataframe: pandas DataFrame containing the data
    - column_name: name of the column
    
    Returns:
    - unique_values: list of unique values in the specified column
    """
    unique_values = dataframe[column_name].unique().tolist()
    return unique_values
```

```python
conditions = [
    (master_df['FIGHTER'] == master_df['FIGHTER_1']) & (master_df['OUTCOME'] == 'W/L'),
    (master_df['FIGHTER'] == master_df['FIGHTER_1']) & (master_df['OUTCOME'] == 'L/W'),
    (master_df['FIGHTER'] == master_df['FIGHTER_1']) & (master_df['OUTCOME'] == 'D/D'),
    (master_df['FIGHTER'] == master_df['FIGHTER_1']) & (master_df['OUTCOME'] == 'NC/NC'),
    (master_df['FIGHTER'] == master_df['FIGHTER_2']) & (master_df['OUTCOME'] == 'W/L'),
    (master_df['FIGHTER'] == master_df['FIGHTER_2']) & (master_df['OUTCOME'] == 'L/W'),
    (master_df['FIGHTER'] == master_df['FIGHTER_2']) & (master_df['OUTCOME'] == 'D/D'),
    (master_df['FIGHTER'] == master_df['FIGHTER_2']) & (master_df['OUTCOME'] == 'NC/NC')
]

choices = ['Win', 'Loss', 'Draw', 'No Contest', 'Loss', 'Win', 'Draw', 'No Contest']

master_df['RESULT'] = np.select(conditions, choices, default=np.nan)
```

```python
# Convert 'DOB' column to datetime
master_df['DOB'] = pd.to_datetime(master_df['DOB'], errors='coerce')

master_df['DATE'] = pd.to_datetime(master_df['DATE'], errors='coerce')


# Calculate the age by subtracting 'DOB' from 'DATE' and dividing by timedelta of 365.25 days per year
master_df['AGE'] = (master_df['DATE'] - master_df['DOB']).dt.days // 365
```

```python
# Group by WEIGHTCLASS and calculate the mean of HEIGHT for each group
master_df['HEIGHT_AVG'] = master_df.groupby('WEIGHTCLASS')['HEIGHT'].transform('mean')

# Group by WEIGHTCLASS and calculate the mean of REACH for each group
master_df['REACH_AVG'] = master_df.groupby('WEIGHTCLASS')['REACH'].transform('mean')

# Group by WEIGHTCLASS and calculate the mean of AGEfor each group
master_df['AGE_AVG'] = master_df.groupby('WEIGHTCLASS')['AGE'].transform('mean')

# Group by WEIGHTCLASS and calculate the mean of APE_INDEX for each group
master_df['AI_AVG'] = master_df.groupby('WEIGHTCLASS')['APE_INDEX'].transform('mean')

# Group by WEIGHTCLASS and calculate the mean of FRAME for each group
master_df['FRAME_AVG'] = master_df.groupby('WEIGHTCLASS')['FRAME'].transform('mean')


# HEIGHT Difference with HEIGHT_MEAN

# Create a new column 'cumulative_avg' that contains the cumulative average for each group up to the current date
# master_df['HEIGHT_MEAN'] = master_df.groupby('WEIGHTCLASS')['HEIGHT'].expanding().mean().reset_index(level=0, drop=True)
```

```python
#Metric Differentials

master_df['HEIGHT_DIFF'] = master_df['HEIGHT'] - master_df['HEIGHT_AVG']

master_df['REACH_DIFF'] = master_df['REACH'] - master_df['REACH_AVG']

master_df['AGE_DIFF'] = master_df['AGE'] - master_df['AGE_AVG']

master_df['AI_DIFF'] = master_df['APE_INDEX'] - master_df['AI_AVG']

master_df['FRAME_DIFF'] = master_df['FRAME'] - master_df['FRAME_AVG']
```
