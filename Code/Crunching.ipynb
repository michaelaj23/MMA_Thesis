```python
def unique_column_values(dataframe, column_name):
    """
    Returns a list of unique values in the specified column of a DataFrame.
    
    Args:
    - dataframe: pandas DataFrame containing the data
    - column_name: name of the column
    
    Returns:
    - unique_values: list of unique values in the specified column
    """
    unique_values = dataframe[column_name].unique().tolist()
    return unique_values
```

```python
# Create a subset of the dataset containing only 'EVENT', 'FIGHTER', 'DATE', and 'Buyrate' and remove duplicates
# Assuming 'DATE' is a column in the dataset. If it's not found, this code will raise an error.
fighter_buyrate_df = df[['EVENT', 'FIGHTER', 'DATE', 'Buyrate']].drop_duplicates()

# Display the first few rows to verify the subset with no duplicates, including 'DATE'
fighter_buyrate_df
```

```python
# Group the original dataset by 'FIGHTER' and sum the 'Buyrate' for each fighter
fighter_buyrate_crunch_df = fighter_buyrate_df.groupby('FIGHTER')['Buyrate'].sum().reset_index()

# Create a 'Rank' column based on the descending order of 'Buyrate' sum
# fighter_buyrate_crunch_df['Buyrate_Rank'] = fighter_buyrate_crunch_df['Buyrate'].rank(method='dense', ascending=False)

# Sort the dataframe based on 'Rank' from first to last (i.e., lowest to highest rank)
# fighter_buyrate_crunch_df = fighter_buyrate_crunch_df.sort_values(by='Buyrate_Rank')

# Display the sorted dataframe
fighter_buyrate_crunch_df
```

```python
# Calculate the average 'Buyrate' by 'EVENT' for each fighter
avg_buyrate_per_event = fighter_buyrate_df.groupby(['FIGHTER', 'EVENT'])['Buyrate'].mean().groupby(level=0).mean().reset_index(name='Avg_Buyrate_Per_Event')

# Merge this with the buyrate_crunch_df dataframe
fighter_buyrate_crunch_df = pd.merge(fighter_buyrate_crunch_df, avg_buyrate_per_event, on='FIGHTER')


# Create a 'Rank_Avg_Buyrate' column for this average 'Buyrate'
# fighter_buyrate_crunch_df['Rank_Avg_Buyrate'] = fighter_buyrate_crunch_df['Avg_Buyrate_Per_Event'].rank(method='dense', ascending=False)

# Sort the dataframe based on 'Rank_Avg_Buyrate'
# buyrate_crunch_df = buyrate_crunch_df.sort_values(by='Rank_Avg_Buyrate')

# Create new column that divide Buyrate by Avg
fighter_buyrate_crunch_df['PPV_Fights'] = fighter_buyrate_crunch_df['Buyrate'] / fighter_buyrate_crunch_df['Avg_Buyrate_Per_Event']


# Display the updated dataframe
fighter_buyrate_crunch_df
```

```python
fighter_buyrate_crunch_df['Buyrate_PERCENTILE'] = (fighter_buyrate_crunch_df['Buyrate'].rank(pct=True)) * 100

fighter_buyrate_crunch_df['PPV_Fights_PERCENTILE'] = (fighter_buyrate_crunch_df['PPV_Fights'].rank(pct=True)) * 100

# fighter_buyrate_crunch_df = fighter_buyrate_crunch_df[df['Buyrate'] > 0]
```

```python
# Group by FIGHTER and aggregate counts based on specified conditions
fighter_stats_df = filtered_df.groupby('FIGHTER').agg(
    Total_Fights=pd.NamedAgg(column='FIGHT_ID', aggfunc='count'),
    Wins=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: (filtered_df.loc[x.index, 'RESULT'] == 'Win').sum()),
    Losses=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: (filtered_df.loc[x.index, 'RESULT'] == 'Loss').sum()),
    Draws=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: (filtered_df.loc[x.index, 'RESULT'] == 'Draw').sum()),
    No_Contests=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: (filtered_df.loc[x.index, 'RESULT'] == 'No Contest').sum()),
    Win_by_KO_TKO=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: ((filtered_df.loc[x.index, 'RESULT'] == 'Win') & (filtered_df.loc[x.index, 'METHOD1'] == 'KO/TKO')).sum()),
    Win_by_Submission=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: ((filtered_df.loc[x.index, 'RESULT'] == 'Win') & (filtered_df.loc[x.index, 'METHOD1'] == 'Submission')).sum()),
    Win_by_Decision=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: ((filtered_df.loc[x.index, 'RESULT'] == 'Win') & (filtered_df.loc[x.index, 'METHOD1'] == 'Decision')).sum()),
    Loss_by_KO_TKO=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: ((filtered_df.loc[x.index, 'RESULT'] == 'Loss') & (filtered_df.loc[x.index, 'METHOD1'] == 'KO/TKO')).sum()),
    Loss_by_Submission=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: ((filtered_df.loc[x.index, 'RESULT'] == 'Loss') & (filtered_df.loc[x.index, 'METHOD1'] == 'Submission')).sum()),
    Loss_by_Decision=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: ((filtered_df.loc[x.index, 'RESULT'] == 'Loss') & (filtered_df.loc[x.index, 'METHOD1'] == 'Decision')).sum())
).reset_index()

fighter_stats_df
```

```python
# Recreate the aggregation for the filtered DataFrame
fighter_stats_with_title_df = df_with_title.groupby('FIGHTER').agg(
    Total_Fights_with_Title=pd.NamedAgg(column='FIGHT_ID', aggfunc='count'),
    Wins_with_Title=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: (df_with_title.loc[x.index, 'RESULT'] == 'Win').sum()),
    Losses_with_Title=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: (df_with_title.loc[x.index, 'RESULT'] == 'Loss').sum()),
    Draws_with_Title=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: (df_with_title.loc[x.index, 'RESULT'] == 'Draw').sum()),
    No_Contests_with_Title=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: (df_with_title.loc[x.index, 'RESULT'] == 'No Contest').sum()),
    Win_by_KO_TKO_with_Title=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: ((df_with_title.loc[x.index, 'RESULT'] == 'Win') & (df_with_title.loc[x.index, 'METHOD1'] == 'KO/TKO')).sum()),
    Win_by_Submission_with_Title=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: ((df_with_title.loc[x.index, 'RESULT'] == 'Win') & (df_with_title.loc[x.index, 'METHOD1'] == 'Submission')).sum()),
    Win_by_Decision_with_Title=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: ((df_with_title.loc[x.index, 'RESULT'] == 'Win') & (df_with_title.loc[x.index, 'METHOD1'] == 'Decision')).sum()),
    Loss_by_KO_TKO_with_Title=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: ((df_with_title.loc[x.index, 'RESULT'] == 'Loss') & (df_with_title.loc[x.index, 'METHOD1'] == 'KO/TKO')).sum()),
    Loss_by_Submission_with_Title=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: ((df_with_title.loc[x.index, 'RESULT'] == 'Loss') & (df_with_title.loc[x.index, 'METHOD1'] == 'Submission')).sum()),
    Loss_by_Decision_with_Title=pd.NamedAgg(column='FIGHT_ID', aggfunc=lambda x: ((df_with_title.loc[x.index, 'RESULT'] == 'Loss') & (df_with_title.loc[x.index, 'METHOD1'] == 'Decision')).sum())
).reset_index()

fighter_stats_with_title_df 
```

```python
# Merge the original stats DataFrame with the one for title fights
fighter_stats_combined_df = pd.merge(fighter_stats_df, fighter_stats_with_title_df, on='FIGHTER', how='left').fillna(0)

# Add a column for Win Percentage
fighter_stats_combined_df['Win_%'] = (fighter_stats_combined_df['Wins'] / fighter_stats_combined_df['Total_Fights']) * 100


# Add a column for Loss Percentage
fighter_stats_combined_df['Loss_%'] = (fighter_stats_combined_df['Losses'] / fighter_stats_combined_df['Total_Fights']) * 100


# Add a column for Win Percentage by KO
fighter_stats_combined_df['KO_Win_%'] = (fighter_stats_combined_df['Win_by_KO_TKO'] / fighter_stats_combined_df['Total_Fights']) * 100

# Add a column for Win Percentage by Subs
fighter_stats_combined_df['Sub_Win_%'] = (fighter_stats_combined_df['Win_by_Submission'] / fighter_stats_combined_df['Total_Fights']) * 100

# Add a column for Win Percentage by Decision
fighter_stats_combined_df['Dec_Win_%'] = (fighter_stats_combined_df['Win_by_Decision'] / fighter_stats_combined_df['Total_Fights']) * 100


# Add a column for Loss Percentage by KO
fighter_stats_combined_df['KO_Loss_%'] = (fighter_stats_combined_df['Loss_by_KO_TKO'] / fighter_stats_combined_df['Total_Fights']) * 100


# Add a column for Loss Percentage by Subs
fighter_stats_combined_df['Sub_Loss_%'] = (fighter_stats_combined_df['Loss_by_Submission'] / fighter_stats_combined_df['Total_Fights']) * 100

# Add a column for Loss Percentage by Decision
fighter_stats_combined_df['Dec_Loss_%'] = (fighter_stats_combined_df['Loss_by_Decision'] / fighter_stats_combined_df['Total_Fights']) * 100


# Add a column for Title Win Percentage
fighter_stats_combined_df['Title_Win_%'] = (fighter_stats_combined_df['Wins_with_Title'] / fighter_stats_combined_df['Total_Fights_with_Title']) * 100


# Add a column for Title Loss Percentage
fighter_stats_combined_df['Title_Loss_%'] = (fighter_stats_combined_df['Losses_with_Title'] / fighter_stats_combined_df['Total_Fights_with_Title']) * 100


# Add a column for Title Win Percentage by KO
fighter_stats_combined_df['Title_KO_Win_%'] = (fighter_stats_combined_df['Win_by_KO_TKO_with_Title'] / fighter_stats_combined_df['Total_Fights_with_Title']) * 100

# Add a column for Title Win Percentage by Subs
fighter_stats_combined_df['Title_Sub_Win_%'] = (fighter_stats_combined_df['Win_by_Submission_with_Title'] / fighter_stats_combined_df['Total_Fights_with_Title']) * 100

# Add a column for Title Win Percentage by Decision
fighter_stats_combined_df['Title_Dec_Win_%'] = (fighter_stats_combined_df['Win_by_Decision_with_Title'] / fighter_stats_combined_df['Total_Fights_with_Title']) * 100


# Add a column for Title Loss Percentage by KO
fighter_stats_combined_df['Title_KO_Loss_%'] = (fighter_stats_combined_df['Loss_by_KO_TKO_with_Title'] / fighter_stats_combined_df['Total_Fights_with_Title']) * 100


# Add a column for Title Loss Percentage by Subs
fighter_stats_combined_df['Title_Sub_Loss_%'] = (fighter_stats_combined_df['Loss_by_Submission_with_Title'] / fighter_stats_combined_df['Total_Fights_with_Title']) * 100

# Add a column for Title Loss Percentage by Decision
fighter_stats_combined_df['Title_Dec_Loss_%'] = (fighter_stats_combined_df['Loss_by_Decision_with_Title'] / fighter_stats_combined_df['Total_Fights_with_Title']) * 100


# Convert counts from floats to ints where they may have been affected by NaNs from the merge
# int_columns = fighter_stats_combined_df.columns[1:]
# fighter_stats_combined_df[int_columns] = fighter_stats_combined_df[int_columns].astype(int)

```
