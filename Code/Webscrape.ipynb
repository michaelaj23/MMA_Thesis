```python
# Send an HTTP GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table element using its tag
table = soup.find('table')

# Check if the table is found
if table:
    # Find all rows (tr elements) within the table
    rows = table.find_all('tr')

    # Iterate over each row and print its contents
    for row in rows:
        # Find all cells (td or th elements) within the row
        cells = row.find_all(['td', 'th'])
        
        # Extract the text content from each cell and join it with a tab delimiter
        row_data = '\t'.join(cell.get_text(strip=True) for cell in cells)
        
        # Print the row data
        print(row_data)
else:
    print('No table found on the page.')
```

```python

# Check if the table is found
if table:
    # Initialize an empty list to store table data
    data = []
    
    # Find all rows (tr elements) within the table
    rows = table.find_all('tr')

    # Extract the first row as column headers
    header_row = rows[0].find_all(['td', 'th'])
    headers = [cell.get_text(strip=True) for cell in header_row]
    
    # Iterate over each row (excluding the first one) and append its contents to the data list
    for row in rows[1:]:
        # Find all cells (td or th elements) within the row
        cells = row.find_all(['td', 'th'])
        
        # Extract the text content from each cell and append to the data list
        row_data = [cell.get_text(strip=True) for cell in cells]
        data.append(row_data)
    
    # Create a pandas DataFrame from the data list and specify column headers
    df = pd.DataFrame(data, columns=headers)
    
    # Print the DataFrame
    print(df)
else:
    print('No table found on the page.')
```

```python
# Create an empty dataframe to store data
data = pd.DataFrame(columns=['Title', 'Date', 'Total Gate', 'Buyrate'])

# Loop through the range of UFC events
for i in range(26, 40):
    url = f'https://en.wikipedia.org/wiki/UFC_{i}'
    response = requests.get(url)
    if response.status_code == 200:  # Check if the page exists
        soup = BeautifulSoup(response.text, 'html.parser')
        title = soup.find('h1', {'id': 'firstHeading'}).text
        info_box = soup.find('table', {'class': 'infobox'})
        if info_box:
            rows = info_box.find_all('tr')
            date = 'Not available'
            gate = 'Not available'
            buyrate = 'Not available'
            for row in rows:
                header = row.find('th')
                if header and 'Date' in header.text:
                    date = row.find('td').text.strip()
                if header and 'Total gate' in header.text:
                    gate_cell = row.find('td')
                    gate = gate_cell.text.strip() if gate_cell else 'Not available'
                if header and 'Buyrate' in header.text:
                    buyrate_cell = row.find('td')
                    buyrate = buyrate_cell.text.strip() if buyrate_cell else 'Not available'
            data = data.append({'Title': title, 'Date': date, 'Total Gate': gate, 'Buyrate': buyrate}, ignore_index=True)
    else:
        print(f'Page for UFC_{i} not found.')
```
